# посчитаем количество телефонных номеров для каждой гостиницы


```python
def count_valid_phones(platform2, phone_columns):
    """
    Подсчитывает количество действительных телефонных номеров для каждой строки DataFrame,
    исключая 'nan' и 'None'.

    Параметры:
    df : pandas.DataFrame
        DataFrame, в котором необходимо подсчитать телефонные номера.
    phone_columns : list
        Список столбцов, содержащих телефонные номера.

    Возвращает:
    pandas.Series
        Серия с количеством действительных телефонных номеров для каждой строки.
    """
    # Заменяем 'None' и строковое представление 'nan' на np.nan
    platform2 = platform2.replace(['None', 'nan', np.nan], [np.nan, np.nan, np.nan])
    
    # Подсчитываем количество не-null значений в каждой строке
    return platform2[phone_columns].apply(lambda x: x.notnull().sum(), axis=1)

# Пример использования функции:
phone_columns = ['phone1', 'phone2', 'phone3', 'phone4']
platform2['valid_phones_count'] = count_valid_phones(platform2, phone_columns)
print(platform2.describe())
```

```python
# Используем .apply() для применения .notnull() к каждой группе
grouped_non_null_counts = platform2[['title', 'uid', 'valid_phones_count']].sort_values(by='valid_phones_count', ascending=False)

print(grouped_non_null_counts.head(10))
```

# Вывести квадрат координат размером 1км на 1км, где больше всего гостиниц.

```python
# Создание GeoDataFrame
gdf = gpd.GeoDataFrame(platform1, geometry=gpd.points_from_xy(platform1.lon, platform1.lat))

# Определение размера сетки (1 км на 1 км)
# Для этого нужно преобразовать километры в градусы широты/долготы
# Примерное значение для широты: 1 км ≈ 0.009 градусов
grid_size = 0.009

# Округление координат до ближайшего значения сетки
gdf['grid_latitude'] = (gdf.geometry.y // grid_size) * grid_size
gdf['grid_longitude'] = (gdf.geometry.x // grid_size) * grid_size

# Группировка по сетке и подсчет количества гостиниц в каждом квадрате
hotel_density = gdf.groupby(['grid_latitude', 'grid_longitude']).size().reset_index(name='count')

# Нахождение квадрата с максимальным количеством гостиниц
max_density_square = hotel_density.loc[hotel_density['count'].idxmax()]

print(f"Квадрат с наибольшим количеством гостиниц находится в координатах: {max_density_square['grid_latitude']}, {max_density_square['grid_longitude']} и содержит {max_density_square['count']} гостиниц.")
```
